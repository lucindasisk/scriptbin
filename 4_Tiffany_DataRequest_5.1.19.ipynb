```python
import pandas as pd
from datetime import date
today=str(date.today())
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
tp='T3'
home='/Volumes/iang/ELS_data/ELS_DTI_Analysis/Results'
tanhome='/Volumes/iang/ELS_data/ELS_DTI_Analysis/Tanner_Files'
date_curated='2019-05-03'
```

```python
unc=pd.read_csv(home+'/{}/ELS_{}_Uncinate_Curated_{}.csv'.format(tp, tp, date_curated),header=0, index_col='subject ID').drop(columns=['Age.at.Scan.{}'.format(tp), 'Date.of.Scan.{}'.format(tp)])
cst=pd.read_csv(home+'/{}/ELS_{}_Corticospinal_Curated_{}.csv'.format(tp, tp, date_curated),header=0, index_col='subject ID').drop(columns=['Age.at.Scan.{}'.format(tp), 'Date.of.Scan.{}'.format(tp)])
cc=pd.read_csv(home+'/{}/ELS_{}_Corpus_Callosum_Curated_{}.csv'.format(tp, tp, date_curated),header=0, index_col='subject ID')
cing=pd.read_csv(home+'/{}/ELS_{}_Cingulum_Cingulate_Curated_{}.csv'.format(tp, tp, date_curated),header=0, index_col='subject ID').drop(columns=['Age.at.Scan.{}'.format(tp), 'Date.of.Scan.{}'.format(tp)])
ifof=pd.read_csv(home+'/{}/ELS_{}_IFOF_Curated_{}.csv'.format(tp, tp, date_curated),header=0, index_col='subject ID').drop(columns=['Age.at.Scan.{}'.format(tp), 'Date.of.Scan.{}'.format(tp)])
```

```python
#Tract Length
ids = pd.read_csv(home + '/{}/{}_IDs.csv'.format(tp,tp), header=0).set_index('indexnum') 
ccmaj=pd.read_csv(home + '/{}/{}_CCmajor_Tractlength.csv'.format(tp,tp), header=None).rename(columns={0:'indexnum', 1:'{}.CC.major.length'.format(tp)}).set_index('indexnum')    
ccmin=pd.read_csv(home + '/{}/{}_CCminor_Tractlength.csv'.format(tp,tp), header=None).rename(columns={0:'indexnum', 1:'{}.CC.minor.length'.format(tp)}).set_index('indexnum') 
cgcl=pd.read_csv(home + '/{}/{}_Left_Cingulum_Cingulate_Tractlength.csv'.format(tp,tp), header=None).rename(columns={0:'indexnum',1:'{}.L.CGC.length'.format(tp)}).set_index('indexnum') 
cgcr=pd.read_csv(home + '/{}/{}_Right_Cingulum_Cingulate_Tractlength.csv'.format(tp,tp), header=None).rename(columns={0:'indexnum',1:'{}.R.CGC.length'.format(tp)}).set_index('indexnum') 
cstl=pd.read_csv(home + '/{}/{}_Left_Corticospinal_Tractlength.csv'.format(tp,tp), header=None).rename(columns={0:'indexnum',1: '{}.L.CST.length'.format(tp)}).set_index('indexnum') 
cstr=pd.read_csv(home + '/{}/{}_Right_Corticospinal_Tractlength.csv'.format(tp,tp), header=None).rename(columns={0:'indexnum', 1:'{}.R.CST.length'.format(tp)}).set_index('indexnum') 
ifofl=pd.read_csv(home + '/{}/{}_Left_IFOF_Tractlength.csv'.format(tp,tp), header=None).rename(columns={0:'indexnum', 1:'{}.L.IFOF.length'.format(tp)}).set_index('indexnum') 
ifofr=pd.read_csv(home + '/{}/{}_Right_IFOF_Tractlength.csv'.format(tp,tp), header=None).rename(columns={0:'indexnum', 1:'{}.R.IFOF.length'.format(tp)}).set_index('indexnum') 
ufl=pd.read_csv(home + '/{}/{}_Left_Uncinate_Tractlength.csv'.format(tp,tp), header=None).rename(columns={0:'indexnum', 1:'{}.L.UF.length'.format(tp)}).set_index('indexnum') 
ufr=pd.read_csv(home + '/{}/{}_Right_Uncinate_Tractlength.csv'.format(tp,tp), header=None).rename(columns={0:'indexnum', 1:'{}.R.UF.length'.format(tp)}).set_index('indexnum') 
#Concat dfs
length = pd.concat([ids,ccmaj,ccmin,cgcl,cgcr,cstl,cstr,ifofl,ifofr,ufl,ufr],axis=1,join='outer')
length['subject ID'] = length['subject ID'].astype(str)

#merge motion
mot=pd.read_csv(home + '/{}/{}_motion_LS_5.1.19.csv'.format(tp, tp), header=0,dtype=str)
mota = pd.merge(length, mot, on='subject ID', how='outer')
mota
```

```python
##Read in Tanner data to add
tanner = pd.read_csv(tanhome+'/{}_Tanner_withRescans.csv'.format(tp))
tanner['subject ID'] = tanner['subject ID'].astype(str)
mot2 = pd.merge(tanner,mota, on='subject ID', how='outer')
```

```python
final1=pd.concat([cc,cing,cst,ifof,unc], axis=1, join='outer')
if tp == 'Tmid':
    final1.reset_index(inplace=True)
    final1['subject ID'] = final1['subject ID'].astype(str)
    final1.rename(columns={'index':'subject ID'},inplace=True)
    final=pd.merge(final1, mot2, on='subject ID', how='left')
    final.columns=final.columns.str.replace('Callosum_Forceps_Minor','CC.Minor').str.replace('Callosum_Forceps_Major','CC.Major').str.replace('Left','L').str.replace('Right','R')
    final.columns=final.columns.str.replace('Cingulum_Cingulate', 'CGC').str.replace('Corticospinal', 'CST').str.replace('Uncinate', 'UF').str.replace('subject ID', 'ID').str.replace(' ', '.')
    final.to_csv(home+'/{}/TH_DTI_DataRequest_ELS_{}_{}.csv'.format(tp,tp,today), index=False)
else:    
    final1.reset_index(inplace=True)
    final1['subject ID'] = final1['subject ID'].astype(str)
    final=pd.merge(final1, mot2, on='subject ID', how='left')
    final.columns=final.columns.str.replace('Callosum_Forceps_Minor','CC.Minor').str.replace('Callosum_Forceps_Major','CC.Major').str.replace('Left','L').str.replace('Right','R')
    final.columns=final.columns.str.replace('Cingulum_Cingulate', 'CGC').str.replace('Corticospinal', 'CST').str.replace('Uncinate', 'UF').str.replace('subject ID', 'ID').str.replace(' ', '.')
    final.to_csv(home+'/{}/TH_DTI_DataRequest_ELS_{}_{}.csv'.format(tp,tp,today), index=False)
```
