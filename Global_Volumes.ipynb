```python
from os.path import join
from nipype import Node
from nipype.interfaces.fsl import ImageStats
from pandas import Series, DataFrame
from glob import glob
from datetime import date
```

```python
home = '/Volumes/group/active/BABIES/BABIES_ibeat/Segmentation/Segmented_IdeaLab'
data=[]
dest = '/Volumes/group/active/BABIES/BABIES_ibeat/data_output'
subject_list = glob(home + '/*_GMWM.nii.gz')
today = today = str(date.today())
```

```python
def extract_vals(infile):
    sub = infile.replace(home + '/', '').replace('_GMWM.nii.gz', '')
    sep = ','
    
    fs_wm = Node(ImageStats(in_file = infile,
                       op_string = "-l 249 -u 251 -V"),
             name = 'fs_wm')
    fs_wm.run()
    wm_out1 = fs_wm.result.outputs
    wm_out = str(wm_out1).replace('\nout_stat = [','').split(sep, 1)[0]
    
    fs_gm = Node(ImageStats(in_file = infile,
                       op_string = "-l 149 -u 151 -V"),
             name = 'fs_gm')
    fs_gm.run()
    gm_out1 = fs_gm.result.outputs
    gm_out = str(gm_out1).replace('\nout_stat = [','').split(sep, 1)[0]
    
    fs_all = Node(ImageStats(in_file = infile,
                       op_string = "-l 149 -u 251 -V"),
             name = 'fs_gm')
    fs_all.run()
    all_out1 = fs_all.result.outputs
    all_out = str(all_out1).replace('\nout_stat = [','').split(sep, 1)[0]
    
    datasub = (sub, wm_out, gm_out, all_out)
    data.append(datasub)
```

```python
for infile in subject_list:
    extract_vals(infile)
```

```python
voxdf = DataFrame(data, columns = ['SUBID' , 'WM_Voxel_Count_mm^3', 'GM_Voxel_Count_mm^3' , 'GM+WM_Voxel_Count (mm^3)']).sort_values(by=['SUBID'])
voxdf.to_csv(dest + '/6mo_GM+WM_CorticalVolume'+today+'.csv', index=False)
```
